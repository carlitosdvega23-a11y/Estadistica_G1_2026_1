## Fundamentos de Probabilidad en Ingeniería

En la ingeniería, la probabilidad es el lenguaje matemático esencial para cuantificar y gestionar la incertidumbre. Ya sea en la ingeniería civil (estimando la vida útil de estructuras), agrícola (prediciendo rendimientos de cultivos bajo diferentes condiciones climáticas) o agroindustrial (controlando la calidad de los procesos de producción), comprender la probabilidad nos permite tomar decisiones informadas frente a la variabilidad intrínseca de los sistemas.

### Conceptos Fundamentales

*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no puede predecirse con certeza antes de su realización, pero del cual se conocen todos los posibles resultados (e.g., la inspección de una viga, el lanzamiento de un dado).
*   **Espacio Muestral ($\Omega$):** Es el conjunto de todos los posibles resultados de un experimento aleatorio. Por ejemplo, al inspeccionar la calidad de un producto, el espacio muestral podría ser $\Omega = \{ \text{Aprobado, Fallo Menor, Fallo Crítico} \}$.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Es decir, uno o más resultados específicos de interés. Por ejemplo, el evento $A = \{ \text{Fallo Crítico} \}$ es que la viga o lote inspeccionado tenga un fallo de ese tipo.

### Ecuaciones de Probabilidad

La probabilidad de un evento puede ser determinada de forma teórica o empírica:

*   **Probabilidad Teórica (Regla de Laplace):** Cuando todos los resultados del espacio muestral son igualmente probables, la probabilidad de un evento $A$ se calcula como:
    $$
P(A) = \frac{\text{Número de casos favorables a } A}{\text{Número total de casos posibles}}
    $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Basada en la observación y recolección de datos de un experimento repetido un gran número de veces. Se calcula como:
    $$
P(A) \approx \frac{\text{Número de veces que ocurre } A}{\text{Número total de repeticiones del experimento}}
    $$

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia en Ingeniería

Tras calcular las probabilidades empíricas generales, surge una pregunta crucial en ingeniería: ¿Qué ocurre cuando la ocurrencia de un evento afecta la probabilidad de otro? Por ejemplo, en la ingeniería civil, ¿la probabilidad de un fallo crítico en una estructura de concreto depende del tipo de proveedor del cemento? En la agroindustria, ¿la probabilidad de que un lote de productos sea defectuoso está ligada al turno de producción? Responder a estas preguntas nos permite identificar relaciones causales y tomar decisiones más precisas para mitigar riesgos o mejorar procesos.

### Conceptos Clave

*   **Probabilidad Condicional:** Es la probabilidad de que un evento $A$ ocurra, *dado que* otro evento $B$ ya ha ocurrido. Nos permite actualizar nuestra estimación de incertidumbre basada en nueva información. Se denota como $P(A|B)$.

*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En un contexto de control de calidad, si la probabilidad de un defecto no cambia independientemente del proveedor, entonces el defecto y el proveedor son eventos independientes. Cuando los eventos son independientes, el conocimiento de uno no aporta información sobre el otro.

### Ecuaciones Clave

*   **Probabilidad Condicional:**
    $$
P(A|B) = \frac{P(A \cap B)}{P(B)}
    $$
    Donde $P(A \cap B)$ es la probabilidad de que los eventos $A$ y $B$ ocurran simultáneamente, y $P(B)$ es la probabilidad de que ocurra el evento $B$.

*   **Condición de Independencia:**
    Los eventos $A$ y $B$ son independientes si y solo si:
    $$
P(A|B) = P(A)
    $$
    O, equivalentemente (y a menudo más útil para cálculos):
    $$
P(A \cap B) = P(A) \cdot P(B)
    $$

## Probabilidad Condicional e Independencia en Ingeniería

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
